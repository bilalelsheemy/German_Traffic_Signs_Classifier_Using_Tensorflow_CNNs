# The following Data preproccessing steps are performed:
1- Converting the images to the grayscale as this helps in increasing the accuracy.
2- Data Normalization.
3- Histogram equalization.
4- Data shuffling.

Note that: I've looked for the most vital preprocessing steps fro the traffic signs data sets and I selected the mentioned steps then, but there're lot of other steps left for ex. Data Augmentation (Flipping, rotation,..etc), but these steps sufficed the project requirements.

```python
from sklearn import preprocessing
from sklearn.utils import shuffle
from skimage import exposure
import cv2

### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 

def preprocess_dataset(X, y = None):
    
    # 1- Convert to grayscale, e.g. single Y channel
    X_trainGray = 0.299 * X[:, :, :, 0] + 0.587 * X[:, :, :, 1] + 0.114 * X[:, :, :, 2]
    
    # 2- Scale features to be in [0, 1]
    X_trainNorm = (X_trainGray / 255.).astype(np.float32)
      
    # 3- Apply localized histogram localization
    X_trainHistEq = np.zeros_like(X_trainNorm)
    for i in range(X_trainNorm.shape[0]):
        X_trainHistEq[i] = exposure.equalize_adapthist(X_trainNorm[i])
        
    if y is not None:  
         
        # 4-  Shuffle the data
        X_shuffle, y_shuffle = shuffle(X_trainHistEq, y)

    # Add a single grayscale channel
    X_trainPrep = X_shuffle.reshape(X_shuffle.shape + (1,)) 
    y_trainPrep = y_shuffle
    
    
    return X_trainPrep, y_trainPrep

    
```

# LeNet Architecture

```python
### LeNet Architecture
import tensorflow as tf



### Feel free to use as many code cells as needed.

from tensorflow.contrib.layers import flatten

def LeNet(x):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(6))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    #  Activation.
    conv1 = tf.nn.relu(conv1)

    # Dropout Layer
    conv1_drpout = tf.layers.dropout(conv1, 0.5)
    
    #  Pooling1 . Input = 28x28x6. Output = 14x14x6.
    pool1 = tf.nn.max_pool(conv1_drpout, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    #  Layer 2: Convolutional.Input = 14x14x6. Output = 10x10x16.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(16))
    conv2   = tf.nn.conv2d(pool1, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b
    
    #  Activation.
    conv2 = tf.nn.relu(conv2)

    
    #  Layer 2': Convolutional.Input = 10x10x16. Output = 10x10x16.
    #conv3_W = tf.Variable(tf.truncated_normal(shape=(10, 10, 16, 16), mean = mu, stddev = sigma))
    #conv3_b = tf.Variable(tf.zeros(16))
    #conv3   = tf.nn.conv2d(conv2, conv3_W, strides=[1, 1, 1, 1], padding='SAME') + conv3_b
       
    # Activation.
    #conv3 = tf.nn.relu(conv3)
    
    # Pooling2 . Input = 10x10x16. Output = 5x5x16.
    pool2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    #  Flatten. Input = 5x5x16. Output = 400.
    fc0   = flatten(pool2)
    
    #  Layer 3: Fully Connected. Input = 400. Output = 120.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(400, 120), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(120))
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    
    #  Activation.
    fc1    = tf.nn.relu(fc1)

    #  Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(120, 84), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(84))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    #  Activation.
    fc2    = tf.nn.relu(fc2)

    #  Layer 5: Fully Connected. Input = 84. Output = 10.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(84, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
```

# VGGNet Architecture

```python
## VGGNet Architecture

import tensorflow as tf


from tensorflow.contrib.layers import flatten

def VGGNet16(x):
    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 30x30x64.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 1, 64), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(64))
    conv1   = tf.nn.conv2d(x, conv1_W, strides=[1, 1, 1, 1], padding='VALID') + conv1_b

    #  Activation.
    conv1 = tf.nn.relu(conv1)
    
    #Droupout Layer
    conv1_drpout = tf.nn.dropout(conv1, 0.5)
    
    #print(conv1.get_shape())
    #  Layer 2: Convolutional. Input = 30x30x64 . Output = 28x28x64.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 64, 64), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(64))
    conv2   = tf.nn.conv2d(conv1_drpout, conv2_W, strides=[1, 1, 1, 1], padding='VALID') + conv2_b

    #  Activation.
    conv2 = tf.nn.relu(conv2)

    #print(conv2.get_shape())
    #  Pooling1. Input = 28x28x64. Output = 14x14x64.
    pool1 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    #print(pool1.get_shape())
    # Layer 3: Convolutional. Input = 14x14x64. Output = 12x12x128.
    conv3_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 64, 128), mean = mu, stddev = sigma))
    conv3_b = tf.Variable(tf.zeros(128))
    conv3   = tf.nn.conv2d(pool1, conv3_W, strides=[1, 1, 1, 1], padding='VALID') + conv3_b
    
    #  Activation.
    conv3 = tf.nn.relu(conv3)
    
    # Layer 4: Convolutional. Input = 12x12x128. Output = 10x10x128.
    conv4_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 128, 128), mean = mu, stddev = sigma))
    conv4_b = tf.Variable(tf.zeros(128))
    conv4   = tf.nn.conv2d(conv3, conv4_W, strides=[1, 1, 1, 1], padding='VALID') + conv4_b
    
    #  Activation.
    conv4 = tf.nn.relu(conv4)    
    
    #print(conv3.get_shape())
    #  Pooling2. Input = 10x10x128. Output = 5x5x128.
    pool2 = tf.nn.max_pool(conv4, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    
    #print(conv4.shape)
    #  Layer 5: Convolutional. Input = 5x5x128. Output = 3x3x256.
    conv5_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 128, 256), mean = mu, stddev = sigma))
    conv5_b = tf.Variable(tf.zeros(256))
    conv5   = tf.nn.conv2d(pool2, conv5_W, strides=[1, 1, 1, 1], padding='VALID') + conv5_b

    #  Activation.
    conv5 = tf.nn.relu(conv5)
    
    
    #print(conv4.shape)
    #  Layer 6: Convolutional. Input = 3x3x256. Output = 1x1x256.
    conv6_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 256, 256), mean = mu, stddev = sigma))
    conv6_b = tf.Variable(tf.zeros(256))
    conv6   = tf.nn.conv2d(conv5, conv6_W, strides=[1, 1, 1, 1], padding='VALID') + conv6_b

    #  Activation.
    conv6 = tf.nn.relu(conv6)

    #print(conv5.shape)
    #  Pooling3. Input = 1x1x256. Output = 1x1x256.
    #pool3 = tf.nn.max_pool(conv6, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    #print(pool3.shape)
    # SOLUTION: Flatten. Input = 1x1x256. Output = 256.
    fl0   = flatten(conv6)
    
    # SOLUTION: Layer 3: Fully Connected. Input = 256. Output = 4096.
    fc1_W = tf.Variable(tf.truncated_normal(shape=(256, 4096), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(4096))
    fc1   = tf.matmul(fl0, fc1_W) + fc1_b
    
    # SOLUTION: Activation.
    fc1    = tf.nn.relu(fc1)

    # SOLUTION: Layer 4: Fully Connected. Input = 4096. Output = 4096.
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(4096, 4096), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(4096))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # SOLUTION: Activation.
    fc2    = tf.nn.relu(fc2)

    # SOLUTION: Layer 5: Fully Connected. Input = 4096. Output = 43.
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(4096, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
```

```python
## Creating placeholder to feed the data to the model through them then

x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, 43)
```

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

```python
### LeNet
### Train your model here.
rate = 0.005

logits_LeNet = LeNet(x)

cross_entropy_LeNet = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits_LeNet)
loss_operation_LeNet = tf.reduce_mean(cross_entropy_LeNet)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation_LeNet = optimizer.minimize(loss_operation_LeNet)
```

```python
### VGGNet16
### Train your model here.
rate = 0.001


logits_VGGNet16 = VGGNet16(x)

cross_entropy_VGGNet16 = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits_VGGNet16)
loss_operation_VGGNet16 = tf.reduce_mean(cross_entropy_VGGNet16)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation_VGGNet16 = optimizer.minimize(loss_operation_VGGNet16)
```

```python
### LeNet Evaluation
correct_prediction_LeNet = tf.equal(tf.argmax(logits_LeNet, 1), tf.argmax(one_hot_y, 1))
accuracy_operation_LeNet = tf.reduce_mean(tf.cast(correct_prediction_LeNet, tf.float32))
saver = tf.train.Saver()

def evaluate_LeNet(X_data, y_data, BATCH_SIZE):
    num_examples = len(X_data)
    total_accuracy_LeNet = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy_LeNet = sess.run(accuracy_operation_LeNet, feed_dict={x: batch_x, y: batch_y})
        total_accuracy_LeNet += (accuracy_LeNet * len(batch_x))
    return total_accuracy_LeNet / num_examples
```

```python
### VGGNet16 Evaluation 
correct_prediction_VGGNet16 = tf.equal(tf.argmax(logits_VGGNet16, 1), tf.argmax(one_hot_y, 1))
accuracy_operation_VGGNet16 = tf.reduce_mean(tf.cast(correct_prediction_VGGNet16, tf.float32))
saver = tf.train.Saver()

def evaluate_VGGNet16(X_data, y_data, BATCH_SIZE):
    num_examples = len(X_data)
    total_accuracy_VGGNet16 = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy_VGGNet16 = sess.run(accuracy_operation_VGGNet16, feed_dict={x: batch_x, y: batch_y})
        total_accuracy_VGGNet16 += (accuracy_VGGNet16 * len(batch_x))
    return total_accuracy_VGGNet16 / num_examples
```

```python
### LeNet Training
BATCH_SIZE = 128
EPOCHS = 25
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_trainPrep, y_trainPrep)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_trainPrep[offset:end], y_trainPrep[offset:end]
            sess.run(training_operation_LeNet, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate_LeNet(X_validPrep, y_validPrep, BATCH_SIZE)
        print("EPOCH {} ...".format(i+1))
        print("LeNet Validation Accuracy = {:.3f} %".format(validation_accuracy*100))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
```

```python
### VGGNet16 Training 
BATCH_SIZE = 64
EPOCHS = 20
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_trainPrep, y_trainPrep)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_trainPrep[offset:end], y_trainPrep[offset:end]
            sess.run(training_operation_VGGNet16, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate_VGGNet16(X_validPrep, y_validPrep, BATCH_SIZE)
        print("EPOCH {} ...".format(i+1))
        print("VGGNet16 Validation Accuracy = {:.3f} %".format(validation_accuracy*100))
        print()
        
    saver.save(sess, './vggnet')
    print("Model saved")
```

# Conclusion

- As we observe that LeNet & VGGNet models don't perform well in classifying the following classes:

1- Stop Class - Class ID 14.
2- Pedestrians Class - Class ID 27.
3- Speed Limit (60 KM/Hr) - Class ID 3.

- And this is because these classes representations in the training set is relatively low as shown in the data visualization section above, hence the models are not trained enough to classify such classes.



- Also another observation:

Even though VGGNet model failed to classify Pedestrian and Speed Limit-60 signs but it was certain in his worng classification unlike to the LeNet Model which was uncertain in the wrong classifications, I believe that this reflects each model (Precision & Accuracy) so we can roughly say that: 
* VGGNet Model is more precise than LeNet Model (More Certain in the classification even in the wrong ones).
* LeNet Model is more accurate than VGGNet Model (Has more accuracy in the classification).
